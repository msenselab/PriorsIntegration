## Repeated-measures ANOVA on the reproduction difference between IR and BR 

```python
m_sumexpSub_diff = pd.read_csv('generated/sumexpSub_diff_range.csv')
m_sumexpSub_diff.head()
```

```python
# A one tailed paired t-test on the absolute NMRD between Experiment (Exp.1 vs. Exp.2) revealed absolute NMRD in Exp.1 was less than that in Exp.2 [t(318) = -2.969 , p <.01, BF10 = 16.107].
pg.pairwise_ttests(data = m_sumexpSub_diff, 
            dv = 'm_diff_RP_Normalized',  between= ['Exp'], subject = 'NSub', padjust = 'holm', alternative="less").round(3)
```

```python
# A one tailed paired t-test on the absolute NMRD between Experiment (Exp.1 vs. Exp.2) revealed absolute NMRD in Exp.1 was less than that in Exp.2 [t(318) = -2.969 , p <.01, BF10 = 16.107].
pg.pairwise_ttests(data = sumexpSub_diff, 
            dv = 'diff_RP_Normalized',  between= ['Exp'], subject = 'NSub', padjust = 'holm', alternative="less").round(3)
```
